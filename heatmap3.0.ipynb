```python
import pandas as pd
record_id = "18174570"
filename = "Vietnam_Data_cleaned.csv"
url = f"https://zenodo.org/records/{18174570}/files/{filename}?download=1"
df_viet = pd.read_csv(url)

df_viet
```

```python
import pandas as pd
import json
import plotly.express as px
import requests

record_id = "18174570"
filename = "vn.json"
url = f"https://zenodo.org/records/{record_id}/files/{filename}?download=1"

response = requests.get(url)
vn_geojson = response.json()


for feature in vn_geojson["features"]:
    feature["properties"]["name"] = feature["properties"]["name"]

df_vn = df_viet.melt(id_vars=['AREA', 'YEAR'], value_vars=['TEMPERATURE'], value_name='Temperature')

vmin = df_vn['Temperature'].min()
vmax = df_vn['Temperature'].max()

fig = px.choropleth_mapbox(
    df_vn,
    geojson=vn_geojson,
    locations='AREA',
    featureidkey="properties.name",
    color='Temperature',
    animation_frame='YEAR',
    color_continuous_scale="YlOrRd",
    mapbox_style="carto-positron",
    zoom=4,
    center={"lat": 16.5, "lon": 108},
    range_color=[vmin, vmax],
    opacity=0.8)

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
record_id = "18174570"
filename = "China_Data_cleaned.csv"
url = f"https://zenodo.org/records/{18174570}/files/{filename}?download=1"
df_chi = pd.read_csv(url)

df_chi
```

```python
record_id = "18174570"
filename = "china.json"
url = f"https://zenodo.org/records/{record_id}/files/{filename}?download=1"

response = requests.get(url)
china_geojson = response.json()

for feature in china_geojson["features"]:
    props = feature["properties"]
    feature["properties"]["name"] = props.get("NAME_2", props.get("NAME_1"))

df_ch = df_chi.melt(
    id_vars=['AREA', 'YEAR'],
    value_vars=['TEMPERATURE'],
    value_name='Temperature'
)

vmin = df_ch['Temperature'].min()
vmax = df_ch['Temperature'].max()

fig = px.choropleth_mapbox(
    df_ch,
    geojson=china_geojson,
    locations='AREA',
    featureidkey='properties.name',
    color='Temperature',
    animation_frame='YEAR',
    color_continuous_scale='YlOrRd',
    mapbox_style='carto-positron',
    zoom=3,
    center={"lat":35.5, "lon":103},
    range_color=[vmin, vmax],
    opacity=0.8
)

fig.update_layout(margin=dict(r=0, t=0, l=0, b=0))
fig.show()
```

```python
fig.write_html("china_temperature_animation.html",include_plotlyjs="cdn",full_html=True)
from google.colab import files
files.download("china_temperature_animation.html")
```

```python
record_id = "18174570"
filename = "Thailand_Data_cleaned.csv"
url = f"https://zenodo.org/records/{18174570}/files/{filename}?download=1"
df_thai = pd.read_csv(url)

df_thai
```

```python
record_id = "18174570"
filename = "thailand.json"
url = f"https://zenodo.org/records/{record_id}/files/{filename}?download=1"

response = requests.get(url)
geojson = response.json()

for feature in geojson["features"]:
    props = feature["properties"]
    if props.get("NAME_1") == "BangkokMetropolis":
        feature["properties"]["name"] = "BangkokMetropolis"

    elif props.get("NAME_2"):
        feature["properties"]["name"] = props["NAME_2"]

df_th = df_thai.melt(
    id_vars=["AREA", "YEAR"],
    value_vars=["TEMPERATURE"],
    value_name="Temperature"
)

vmin = df_th["Temperature"].min()
vmax = df_th["Temperature"].max()

fig = px.choropleth_mapbox(
    df_th,
    geojson=geojson,
    locations="AREA",
    featureidkey="properties.name",
    color="Temperature",
    animation_frame="YEAR",
    color_continuous_scale="YlOrRd",
    mapbox_style="carto-positron",
    zoom=4.2,
    center={"lat": 14, "lon": 101},
    range_color=[vmin, vmax],
    opacity=0.85
)

fig.update_layout(
    margin=dict(r=0, t=0, l=0, b=0)
)

fig.show()
```

```python
fig.write_html("thailand_temperature_animation.html",include_plotlyjs="cdn",full_html=True)
from google.colab import files
files.download("thailand_temperature_animation.html")
```

```python
import pandas as pd
record_id = "18174570"
filename = "Indonesia_Data_cleaned.csv"
url = f"https://zenodo.org/records/{18174570}/files/{filename}?download=1"
df_indo = pd.read_csv(url)

df_indo
```

```python
record_id = "18174570"
filename = "indonesia.json"
url = f"https://zenodo.org/records/{record_id}/files/{filename}?download=1"

response = requests.get(url)
indo_geojson = response.json()

def force_city_name(geojson):
    for feature in geojson["features"]:
        props = feature["properties"]
        feature["properties"]["name"] = props["NAME_1"]

force_city_name(indo_geojson)

df_ind = df_indo.melt(id_vars=['AREA', 'YEAR'],value_vars=['TEMPERATURE'],value_name='Temperature')

vmin = df_ind["Temperature"].min()
vmax = df_ind["Temperature"].max()

fig = px.choropleth_mapbox(
    df_ind,
    geojson=indo_geojson,
    locations="AREA",
    featureidkey="properties.name",
    color="Temperature",
    animation_frame="YEAR",
    color_continuous_scale="YlOrRd",
    mapbox_style="carto-positron",
    zoom=4,
    center={"lat":-2.5, "lon":118},
    range_color=[vmin, vmax],
    opacity=0.8)

fig.update_layout(margin=dict(r=0, t=0, l=0, b=0))
fig.show()
```
