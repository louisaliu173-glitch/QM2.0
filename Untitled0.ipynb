```python
import pandas as pd
from ipyleaflet import Map, CircleMarker, WidgetControl
import ipywidgets as widgets
import matplotlib.cm as cm
import matplotlib.colors as colors

# ----------------------------
# 示例 df
# df 已有列: AREA, YEAR, TEMPERATURE, LAT, LON
# ----------------------------
df = pd.DataFrame({
    'AREA': ['Jakarta','Kuala Lumpur','Beijing','Mumbai','Tokyo']*3,
    'YEAR': [2015]*5 + [2016]*5 + [2017]*5,
    'TEMPERATURE': [28,27,22,26,16,29,28,23,27,17,30,29,24,28,18],
    'LAT': [-6.2, 3.14, 39.9, 19.0, 35.7],
    'LON': [106.8, 101.7, 116.4, 72.8, 139.7]
})

# ----------------------------
# 创建地图
# ----------------------------
center = [25, 110]
zoom = 3
m = Map(basemap=basemaps.CartoDB.Positron, center=center, zoom=zoom)

# ----------------------------
# 创建 colormap
# ----------------------------
vmin = df['TEMPERATURE'].min()
vmax = df['TEMPERATURE'].max()
norm = colors.Normalize(vmin=vmin, vmax=vmax)
cmap = cm.get_cmap('YlOrRd')

# ----------------------------
# 创建年份滑块
# ----------------------------
years = sorted(df['YEAR'].unique())
slider = widgets.IntSlider(value=years[0], min=years[0], max=years[-1], step=1, description='Year')

markers = []

def update_map(change):
    # 先移除之前的 marker
    for marker in markers:
        m.remove_layer(marker)
    markers.clear()

    # 筛选当年数据
    year_df = df[df['YEAR']==slider.value]

    # 添加 CircleMarker
    for idx, row in year_df.iterrows():
        color = colors.to_hex(cmap(norm(row['TEMPERATURE'])))
        marker = CircleMarker(location=(row['LAT'], row['LON']),
                              radius=10,
                              color=color,
                              fill_color=color,
                              fill_opacity=0.7,
                              popup=row['AREA'] + ': ' + str(row['TEMPERATURE']))
        m.add_layer(marker)
        markers.append(marker)

slider.observe(update_map, names='value')

# 初始显示
update_map(None)

# 添加滑块到地图上
widget_control = WidgetControl(widget=slider, position='topright')
m.add_control(widget_control)

m
```
