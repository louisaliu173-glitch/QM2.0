```python
import pandas as pd
import numpy as np
from ipyleaflet import Map, CircleMarker, basemaps
from branca.colormap import linear
import ipywidgets as widgets

countries = ['Thailand', 'Malaysia']
df_region = df[df['COUNTRY'].isin(countries)]

map_TM = Map(
    center=[10, 107],
    zoom=4,
    basemap=basemaps.CartoDB.Positron
)

colormap = linear.YlOrRd_09.scale(
    df_region['TEMPERATURE'].min(),
    df_region['TEMPERATURE'].max()
)
colormap.caption = 'temperature'
map_TM.add_control(colormap)

markers = {}
for area, group in df_region.groupby('AREA'):
    row = group.iloc[0]

    marker = CircleMarker(
        location=(row['LATITUDE'], row['LONGITUDE']),
        radius=6,
        fill_opacity=0.8,
        stroke=False
    )
    markers[area] = marker
    map_TM.add_layer(marker)

def update_map(year):
    df_year = df_region[df_region['YEAR'] == year]
    temp_dict = dict(zip(df_year['AREA'], df_year['TEMPERATURE']))

    for area, marker in markers.items():
        if area in temp_dict:
            color = colormap(temp_dict[area])
            marker.color = color
            marker.fill_color = color
        else:
            marker.color = '#cccccc'
            marker.fill_color = '#cccccc'

year_slider = widgets.IntSlider(
    value=int(df_region['YEAR'].min()),
    min=int(df_region['YEAR'].min()),
    max=int(df_region['YEAR'].max()),
    step=1,
    description='Year',
    continuous_update=False
)
widgets.interact(update_map, year=year_slider)

map_TM
```

```python
import pandas as pd
import json
import plotly.express as px

geojson_path = '/content/drive/MyDrive/Colab Notebooks/vn.json'
with open(geojson_path, 'r', encoding='utf-8') as f:
    geojson = json.load(f)
for feature in geojson['features']:
    feature['properties']['name'] = str(feature['properties']['name'])

df_long = df_viet.melt(id_vars=['AREA', 'YEAR'], value_vars=['TEMPERATURE'], value_name='Temperature')

vmin = df_long['Temperature'].min()
vmax = df_long['Temperature'].max()

fig = px.choropleth_mapbox(
    df_long,
    geojson=geojson,
    locations='AREA',
    color='Temperature',
    featureidkey="properties.name",
    animation_frame='YEAR',
    color_continuous_scale="YlOrRd",
    mapbox_style="carto-positron",
    zoom=4,
    center={"lat": 16.5, "lon": 108},
    range_color=[vmin, vmax],
    opacity=0.8)

fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```

```python
fig.write_html("vietnam_temperature_animation.html",
    include_plotlyjs="cdn",
    full_html=True)
from google.colab import files
files.download("vietnam_temperature_animation.html")
```
